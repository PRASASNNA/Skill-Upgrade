# Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load the Dataset

```python
# Load the Titanic dataset
titanic_df = pd.read_csv("Titanic+Data+Set.csv")
titanic_df
```

# Explore the Data

```python
# Display the first few rows of the dataset
titanic_df.head()
```

```python
# Check the data types of each column
titanic_df.info()
```

# Handle Missing Values

```python
# Check for missing values
titanic_df.isnull().sum()
```

```python
# Impute missing values with the mean
titanic_df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)
```

```python
# Handling missing values in 'Embarked' column by imputing with mode
titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0], inplace=True)
```

```python
# Handling missing values in 'Cabin' column by creating a new category 'Unknown'
titanic_df['Cabin'].fillna('Unknown', inplace=True)
```

```python
# Check if there are any missing values left
titanic_df.isnull().sum()
```

```python
titanic_df
```

```python
# Save encoded data
titanic_df.to_csv("titanic.csv", index=False)
```

# 2.Use Python and popular libraries(e.g.,Pandas,Matplotlib,Seaborn) for data manipulation and visualiztion.

# Visualize Data Distributions.

```python
# Visualize the distribution of numerical variables using histograms
titanic_df.hist(figsize=(10, 8))
plt.show()
```

```python
# Visualize the distribution of categorical variables using count plots
sns.countplot(x='Survived', data=titanic_df)
plt.show()
```

# Explore Relationships between Variables

```python
# Explore relationships between numerical variables using scatter plots
sns.scatterplot(x='Age', y='Fare', data=titanic_df)
plt.show()
```

```python
# Explore relationships between numerical and categorical variables using box plots
sns.boxplot(x='Survived', y='Age', data=titanic_df)
plt.show()
```

```python
# Explore relationships between categorical variables using stacked bar plots
sns.countplot(x='Survived', hue='Sex', data=titanic_df)
plt.show()
```

# 3.Generate summary statistics, visualiztions, and insights from the dataset

# Summary Statistics:

```python
# Summary statistics of numerical variables
titanic_df.describe()
```

```python
# Summary statistics of categorical variables
print(titanic_df.describe(include=['object']))
```

```python
# Visualize the distribution of passengers by class
sns.countplot(x='Pclass', data=titanic_df)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
# Visualize the distribution of passengers by gender
sns.countplot(x='Sex', data=titanic_df)
plt.title('Passenger Gender Distribution')
plt.show()
```

```python
# Visualize the distribution of passengers by survival
sns.countplot(x='Survived', data=titanic_df)
plt.title('Survival Count')
plt.show()
```

```python
# Visualize the age distribution of passengers
sns.histplot(data=titanic_df, x='Age', bins=20, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.show()
```

```python
# Visualize the survival distribution by age
sns.histplot(data=titanic_df, x='Age', hue='Survived', bins=20, kde=True, multiple='stack')
plt.title('Survival Distribution by Age')
plt.xlabel('Age')
plt.show()
```

```python
# Visualize the survival distribution by gender
sns.countplot(x='Sex', hue='Survived', data=titanic_df)
plt.title('Survival Distribution by Gender')
plt.show()
```

```python
# Visualize the survival distribution by passenger class
sns.countplot(x='Pclass', hue='Survived', data=titanic_df)
plt.title('Survival Distribution by Passenger Class')
plt.show()
```
